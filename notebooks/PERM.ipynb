## Enumerating Gene Orders

https://rosalind.info/problems/perm/

```python
from itertools import permutations
```

```python
def tuple_to_string(value):
    return ' '.join(map(str, value))

def gene_permutations(value):
    result = []
    count_iterations = 0
    
    list_value = list(range(1, value + 1))
    
    for perm in list(permutations(list_value)):
        count_iterations += 1
        result.append(tuple_to_string(perm))
        
    result.insert(0, count_iterations)
    return result
```

```python
with open("../data/rosalind_perm.txt", "r") as reader:
    value = int(reader.read())

for i in gene_permutations(value):
    print(i)
```
