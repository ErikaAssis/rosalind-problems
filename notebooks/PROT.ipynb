# Translating RNA into Protein

https://rosalind.info/problems/prot/

```python
import re
```

```python
AMINO_ACIDS_MAP = {"UUU" : "F", "CUU" : "L", "AUU" : "I", "GUU" : "V",
                   "UUC" : "F", "CUC" : "L", "AUC" : "I", "GUC" : "V",
                   "UUA" : "L", "CUA" : "L", "AUA" : "I", "GUA" : "V",
                   "UUG" : "L", "CUG" : "L", "AUG" : "M", "GUG" : "V",
                   "UCU" : "S", "CCU" : "P", "ACU" : "T", "GCU" : "A",
                   "UCC" : "S", "CCC" : "P", "ACC" : "T", "GCC" : "A",
                   "UCA" : "S", "CCA" : "P", "ACA" : "T", "GCA" : "A",
                   "UCG" : "S", "CCG" : "P", "ACG" : "T", "GCG" : "A",
                   "UAU" : "Y", "CAU" : "H", "AAU" : "N", "GAU" : "D",
                   "UAC" : "Y", "CAC" : "H", "AAC" : "N", "GAC" : "D",
                   "UAA" : "STOP", "CAA" : "Q", "AAA" : "K", "GAA" : "E",
                   "UAG" : "STOP", "CAG" : "Q", "AAG" : "K", "GAG" : "E",
                   "UGU" : "C", "CGU" : "R", "AGU" : "S", "GGU" : "G",
                   "UGC" : "C", "CGC" : "R", "AGC" : "S", "GGC" : "G",
                   "UGA" : "STOP", "CGA" : "R", "AGA" : "R", "GGA" : "G",
                   "UGG" : "W", "CGG" : "R", "AGG" : "R", "GGG" : "G" 
                   }
```

```python
def rna_into_protain(rna):
    rna = re.findall('...', rna)
    protein_list = []
    
    for codon in rna:
        if codon in AMINO_ACIDS_MAP:
            if AMINO_ACIDS_MAP[codon] == "STOP":
                break
            protein_list.append(AMINO_ACIDS_MAP[codon])
            
    return ''.join(protein_list)
```

```python
with open("../data/rosalind_prot.txt", "r") as reader:
    rna_string = reader.read()

result = rna_into_protain(rna_string)
print(result)
```
